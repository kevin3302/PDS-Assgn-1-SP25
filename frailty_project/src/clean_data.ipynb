```python
import pandas as pd

# Load raw data
raw_data = pd.read_csv("raw_frailty_data.csv")

raw_data.head(10)
```

```python
raw_data.info()
```

```python
# Clean the data
# Convert Frailty to binary (0 for N, 1 for Y)
raw_data["Frailty"] = raw_data["Frailty"].map({'N': 0, 'Y': 1})

raw_data.head(10)
```

```python
# Save cleaned data
raw_data.to_csv("frailty_project/data_clean/clean_frailty_data.csv", index=False)
```
